```python
# Importing libraries
from matplotlib import pyplot as plt
from skimage import data
from skimage.feature import blob_dog, blob_doh, blob_log
from math import sqrt
from skimage.color import rgb2gray
import glob 
import matplotlib.cm
from skimage.io import imread
```

```python
# Commom working directories where the image is
cwd = '/home/bishnu/Desktop/ML/'

# The name of the image file 
image = cwd + 'IMG_1494.JPG'

# Reading file 
file = imread (image, as_gray= True)

# plotting files
plt.imshow (file, cmap= matplotlib.cm.gray)

# showing files 
plt.show ()
```

```python
blobs_log = blob_log(file , max_sigma = 30, num_sigma = 10, threshold = 0.1)
blobs_log [:, 2] = blobs_log [:, 2] * sqrt (2)
numrows = len (blobs_log)
print ("Number of Object detected :", numrows)
```

```python
fig,ax = plt.subplots (1,1 )
plt.imshow (file, cmap = matplotlib.cm.gray)
for blob in blobs_log :
    y, x, r = blob 
    c = plt.Circle ((x, y), r+ 5, color = 'lime', linewidth = 2, fill = False)
    ax.add_patch(c)
    plt.show ()
```
